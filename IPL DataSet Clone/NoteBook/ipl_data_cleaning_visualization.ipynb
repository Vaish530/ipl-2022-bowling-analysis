```python
data=pd.read_csv(r"C:/Users/91628/OneDrive/Documents/_My_projects _Vaishnavi/ipl dataset cleanining/Best Bowling Strike Rate Innings - 2022(1).csv")

df=pd.DataFrame(data)
```

```python
ilocation=df.iloc[[1,6,3,5]]
print(ilocation)
```

```python
sort_values=df.sort_values(["Against","Venue"],ascending=[0,1])
print(sort_values)
```

manipulating the data by wkts +overs sumation and storing in w/O colums here new colmns is created

```python
manipulate_val=df["w/O"]=df["Ov"]+df["Wkts"]
print(manipulate_val)
```

droping the created columns using drop method

```python
df.drop(columns="w/O",inplace=True)
```

checking for duplicated value and replacing nan values with overs mode value

```python
dup=df.duplicated()
df.drop_duplicates(inplace=True)
Mode_val_Ov=df["Ov"].mode()[0]
df["Ov"] = df["Ov"].fillna(Mode_val_Ov)
print(Mode_val_Ov)
#print(df)
```

replacing nan values with mean of wkts value

```python
Mean_val_Wkts=round(df["Wkts"].mean())
df["Wkts"] = df["Wkts"].fillna(Mean_val_Wkts)
my_loc = df[["Ov", "Wkts"]]
print(my_loc) 
```

grouping players and thier total runs in the whole ipl matches and analysibg max runs given player and min runs

```python
#player with overall runs***done
player_run_sum = df.groupby("Player")["Runs"].sum().reset_index()
player_run_sum.columns = ["Player", "Total_Runs"]
max_runs=player_run_sum.loc[player_run_sum["Total_Runs"].idxmax(),"Player"]
min_runs=player_run_sum.loc[player_run_sum["Total_Runs"].idxmin(),"Player"]
print(player_run_sum)
```

players with overall wkts and top 10 wkts givers sorted in ascending order

```python
#player with overall wkts
player_wkts_sum = df.groupby("Player")["Wkts"].sum().reset_index()
player_wkts_sum.columns = ["Player", "Total_Wkts"]
player_wkts_sum=player_wkts_sum.sort_values(by="Total_Wkts",ascending=False)
top_ten=player_wkts_sum.head(10)
print(top_ten)
```

player and there overall overs gone through by using sum

```python
# player and there overall overs gone through
player_overs = df.groupby("Player")["Ov"].sum().reset_index()
player_overs.columns = ["Player", "Total_Overs"]
print(player_overs)
```

the overall strike rate of each player along wuth minium strike rate player

```python
player_SR = df.groupby("Player")["SR"].sum().reset_index()
player_SR.columns = ["Player", "sum SR"]
min_SR=player_SR.sort_values(by="sum SR",ascending=True).head(5)
print(min_SR)
```

players oversand thier runs, wickets sumation

```python
#players runs wickets overs**
player_comparison = df.groupby("Player")[["Runs", "Ov","Wkts"]].sum().reset_index()
print(player_comparison)



```

how many bowlers does each team container

```python
# how many players from each team**
team_group=df.groupby("Against")["Player"].count().reset_index()
team_group.columns=["against","players count"]
print(team_group)
```

teams played in each venue

```python

#venues and teams**
venue_team_summary = df.groupby("Venue")["Against"].unique().reset_index()
venue_team_summary.columns = ["Venue", "Teams_At_Venue"]
print(venue_team_summary)
```

 monthly matches played by teams

```python
# monthly matches **
df["Match Date"]=pd.to_datetime(df["Match Date"],format="%d-%b-%y")
df["Month"]=df["Match Date"].dt.month_name()
month_matches=df["Month"].value_counts().reset_index()
month_matches.columns=["month","match_count"]
print(month_matches)
```

venue and noof times does each team played

```python
#venue and noof times teams played 
venue_against=df.groupby("Venue")["Against"].count().reset_index()
venue_against.columns=["venue","against count"]
print(venue_against)



```

*Exploratory Data Analysis (EDA): Players and Their Total Runs 

Insights:
1)The barchart shows the total runs played by each player in the dataset
2)the highest runs given player is highlighted with red
3)where as lowest runs given player is highlighted in blue
4)other players are teal green color

*By using these interpretation we can easily identify the best player who is giving least runs in the whole season foe against team and the player who is giving more runs to the against team. 
we categorized max and min runs given player by  max_runs and min _runs and grouped the players according to the max_run and min_runs coloring is given

```python
#matplotlib
#1) players and their sum of runs along with max runs blue and min runs red color
colors=["#FF9999" if player==max_runs else "#99CCFF" if player==min_runs else "#B2DFDB"  for player in player_run_sum["Player"]]
plt.figure(figsize=(10,6))
plt.bar(player_run_sum["Player"],player_run_sum["Total_Runs"],color=colors)
plt.title("Player and their sum of runs")
plt.xlabel("Player")
plt.ylabel("Total_Runs")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()
```

* EDA : Player Runs, Wickets, and Overs Comparison
  Insights:
  1) The grouped bar shows the performance of each player by using the terms of runs ,overs,wkts
  2) color-coding helps in differentiate runs,overs and wickets
  3) by these chart we cand easily determine the player given runs,wkts taken and overs played

```python
#4)players runs wickets overs comarision
plt.figure(figsize=(20,10))
players=player_comparison["Player"]
runs=player_comparison["Runs"]
overs=player_comparison["Ov"]
wickets=player_comparison["Wkts"]
x = np.arange(len(players))
width=0.25
plt.bar(x-width,runs,width=width,label="runs",color="#B39DDB")
plt.bar(x,overs,width=width,label="overs",color="#FF8A65")
plt.bar(x+width,wickets,width=width,label="wickets",color="#6EC6FF")
plt.title("players runs wickets overs comarision")
plt.xlabel("Player")
plt.ylabel("total")
plt.xticks(x, players, rotation=45, ha="right")
max_val = max(runs.max(), overs.max(),wickets.max())
plt.yticks(np.arange(0, max_val + 2, 2)) 
plt.legend(loc="upper right",framealpha=0.5,shadow=True,fancybox=True,facecolor="white",edgecolor="black")
plt.tight_layout()
plt.show()
```
